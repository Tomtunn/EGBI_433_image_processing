```python
def get_df_ltitem_hierarchy(o: Any, depth=0):
    """Show location and text of LTItem and all its descendants"""
    df = pd.DataFrame(columns=['element', 'x1', 'y1', 'x2', 'y2', 'text'])

    df = df.append({
        'element': get_indented_name(o, depth),
        'x1': get_optional_bbox(o)[0],
        'y1': get_optional_bbox(o)[1],
        'x2': get_optional_bbox(o)[2],
        'y2': get_optional_bbox(o)[3],
        'text': get_optional_text(o)
    }, ignore_index=True)

    if isinstance(o, Iterable):
        for i in o:
            df_tmp = get_df_ltitem_hierarchy(i, depth=depth + 1)
            df.append(df_tmp, ignore_index=True)

    return df
```
